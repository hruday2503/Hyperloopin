```python
feature_counts=binary_df.sum().sort_values(ascending=False)
```

```python
top5=feature_counts.head(5)
tail5=feature_counts.tail(5)
```

```python
#sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
sns.barplot(x=top5.values, y=top5.index, palette="Blues_d")
plt.title("Top 5 Binary Features (Most Frequent)")
plt.xlabel("Count")
plt.ylabel("measure")
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=tail5.values, y=tail5.index, palette="Reds_d")
plt.title("Bottom 5 Binary Features (Least Frequent)")
plt.xlabel("Count")
plt.ylabel("measure")
plt.tight_layout()
plt.show()
```

```python
def determine_membership(row):
    return (
        row.get("Sensor_Normal", 0) == 1 and
        (row.get("Fault Risk_None", 0) == 1 or row.get("Fault Risk_Low", 0) == 1) and
        (row.get("Battery Level_Medium", 0) == 1 or row.get("Battery Level_High", 0) == 1) and
        (row.get("Power Mode_Normal", 0) == 1 or row.get("Power Mode_Economy", 0) == 1) and
        (row.get("Speed_Medium", 0) == 1 or row.get("Speed_High", 0) == 1) and
        row.get("Temperature_Medium", 0) == 1
    )
```

```python
binary_df['Is_Member'] = binary_df.apply(determine_membership, axis=1)
```

```python
members = binary_df[binary_df['Is_Member'] == True]
print("✅ Qualified Members:")
print(members)
```

```python
members.to_csv("qualifiedmemvers.csv",index=False)
```
